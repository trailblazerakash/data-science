# Indian governmnet data visualization of percentage of textbook get by schools 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset.csv")
```

```python
print(df.head())
```

```python
states=df['State/UTs']
y_pos= np.arange(len(states))
primary = df['Primary Schools 2012-13']
allschools = df['All Schools 2012-13']
```

```python
bar_width=0.6
plt.barh(y_pos, primary,bar_width,alpha=0.5)
plt.yticks(y_pos, states,size=3)
plt.xlabel('primary schools % who get books')
plt.ylabel("states and UT")
plt.title('graph')
plt.figure(figsize=(10,10))
plt.show()
```

```python
plt.barh(y_pos,allschools,color='r')
plt.yticks(y_pos,states,size=3)
plt.xlabel("States/UT")
plt.ylabel("schools % who get books")
plt.show()
```

```python
fig,ax= plt.subplots()
opacity=0.3
width=0.3
rects1 = plt.barh(y_pos,primary,opacity,color='r',label='Primary Schools')
rects2=plt.barh(y_pos+width,allschools,opacity,color='g',label='all schools')
plt.xlabel("% of colleges")
plt.ylabel("states and UT")
plt.yticks(y_pos , states)
plt.legend()
plt.tight_layout()
plt.show()
```
